```python
l = ["This book is very interesting","this book is not good","this novel is written by chetan"]
```

```python
sample = pd.DataFrame(l,columns=["message"])
```

```python
sample["Target"] = ["Positive","negative","positive"]
```

```python
sample
```

```python
vector = CountVectorizer()
vector.fit(sample.message)
```

```python
vector.get_feature_names()
```

```python
sample_vec = vector.transform(sample.message)
```

```python
pd.DataFrame(sample_vec.toarray(),columns=vector.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit(sample.message)
```

```python
tfidf.get_feature_names()
```

```python
tfidf_vec = tfidf.transform(sample.message)
```

```python
pd.DataFrame(tfidf_vec.toarray(),columns=tfidf.get_feature_names())
```
